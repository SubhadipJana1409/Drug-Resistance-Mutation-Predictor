# Day 23 · Drug Resistance Mutation Predictor

Explore SNP profiles and resistance predictions.

```python
import sys; sys.path.insert(0,'..')
from src.data.simulator import simulate_isolates, DRUGS
from src.models.predictor import DrugResistancePredictor
from sklearn.model_selection import train_test_split

X, Y = simulate_isolates(n_per_drug=80)
X_tr, X_te, Y_tr, Y_te = train_test_split(X, Y, test_size=0.2, random_state=42)

pred = DrugResistancePredictor(drugs=DRUGS)
pred.fit(X_tr, Y_tr)
results = pred.evaluate(X_te, Y_te)

for drug in DRUGS:
    auc = results[drug]['logistic_l1']['auc_roc']
    print(f'{drug:<15} AUC={auc:.3f}')
```
